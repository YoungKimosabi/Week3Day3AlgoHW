## Generators <br>
<p>Generators are a type of iterable, like lists or tuples. They do not allow indexing, but they can still be iterated through with for loops. They are created using functions and the yield statement.</p>

```python
def myRange(start, stop, step=2):
    while start<stop:
        yield start #yeild denotes a generator
        start += step
        
x = myRange(4, 20)
print(x)
for i in x:
    print(i)
```

#### Yield Keyword <br>
<p>The yield keyword denotes a generator, it doesn't return so it won't leave the function and reset all variables in the function scope, instead it yields the number back to the caller.</p>

## In-Place Algorithms

#### Syntax

```python
# swap algorithm
def swap(list1, x, y):
    list1[x], list1[y] = list1[y], list1[x]
    return

myList = [4, 20, 10, 5, 13]
print(f'Before swap: {myList}')

output = swap(myList, 1, -1)
print(f' After swap:{myList}')
print('output:', output)
```

#### Out of Place Algorithm

```python
def swap(list1, x, y):
    copy = list1[:]
    copy[x], copy[y] = copy[y], copy[x]
    return copy

myList = [4, 20, 10, 5, 13]
print(f'Before swap: {myList}')

output = swap(myList, 1, -1)
print(f' After swap:{myList}')
print('output:', output)
```

## Two Pointers

#### Syntax

```python
def twoPointer(list1):
    left = 0
    right = len(list1)-1
    while left < right:
        list1[left], list1[right] = list1[right], list1[left]
        left +=1
        right -=1
    return

myList = [1, 2, 3, 4, 5, 6, 7, 8, 9]
twoPointer(myList)
print(myList)
```

# Sorting Algorithms

#### Bubble Sort

Worst Case: O(n^2) Time - O(1) Space

```python
def swap(list1, x, y): 
    list1[x], list1[y] = list1[y], list1[x]
    return

def bubbleSort(nums):
    isSorted = False
    count = 0
    
    while isSorted == False:
        performedSwap = False
        for i in range(len(nums)-1-count):
            #if right num is smaller, then sway
            if nums[i] > nums[i+1]:
                swap(nums, i , i+1)
                performedSwap=True
        if performedSwap == False:
            isSorted = True
        count +=1
        if count == len(nums)-1:
            isSorted = True
            
my_List = [9, 20, 63, 45, 69, 19, 5, 3]
bubbleSort(my_List)
print(my_List)
```

##### Insertion Sort

Worst Case: O(n^2) time - O(1)space

```python
def insertionSort(nums):
    for i in range(1, len(nums)):
        j = i
        while j > 0 and nums[j] < nums[j-1]:
            swap(nums, j, j-1)
            j-=1
my_List = [9, 20, 63, 45, 69, 19, 5, 3]            
insertionSort(my_List)
print(my_List)
```

## Merge Sort

#### How it Works

```python
# step1 split everything into its own grou

# step2 from left to right, merge groups together
# step3 while merging, place each item in the correct position within  the merged group
# step4 continue steps 3-4 until 1 group is left

def mergeSort(nums):
    #step1
    if len(nums)>1:
        mid = len(nums)//2
        lefthalf = nums[:mid]
        righthalf = nums[mid:]
        
        #recursively split left and right halves if needed
        mergeSort(lefthalf)
        mergeSort(righthalf)
        
        
        # index pointer for 3 lists
        i = 0 
        j = 0
        k = 0
        
        #step 2
        while i<len(lefthalf) and j<len(righthalf):
            if lefthalf[i] < righthalf[j]:
                nums[k] = lefthalf[i]
                i += 1
            else:
                nums[k] = righthalf[j]
                j += 1
                
            k+=1
        
        
        #step3 when one side finishes... place the other list into the main array
        while i < len(lefthalf):
            nums[k] = lefthalf[i]
            i+=1
            k+=1
        while j < len(righthalf):
            nums[k] = righthalf[j]
            j+=1
            k+=1
        
        
    return

my_List = [9, 20, 63, 45, 69, 19, 5, 3]
mergeSort(my_List)
print(my_List)
```
